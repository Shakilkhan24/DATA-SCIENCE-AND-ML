```python
mask1=df['country-code'].apply(lambda x:isinstance(x,(int,float)) and x>=16 if not pd.isna(x) else False)

    # isinstance(x,(int,float)) ____CHEACKING x IS 'integer or float'
    # not .isna(x) ____for ignoring 'NaN' values

    # x>=16 __ it is a boolean expresion____returns true or false
            # SO WE CAN USE ANY BOOLEAN FUNCTION there

df[mask1]
```

instead of 'x>=16' you can use '**boolean_func(x)**' there

```python
def boolean_func(x):
  if x>=16:
    return True
  else:
    return False

mask2=df['country-code'].apply(lambda x:isinstance(x,(int,float)) and boolean_func(x) if not pd.isna(x) else False)

# mask1 == mask2

df[mask2]
```

Custom function example___

finding next prime , if non_prime

```python
# checking prime

def prime_check(x):
  if x<2:
    return False
  for i in range(2,int(x**0.5)+1):
    if x%i==0:
      return False
  return True

# if not prime, finding next prime

def findingNextPrime(x):
  if prime_check(x):
    return x
  else:
    while not prime_check(x):
      x+=1
    return x
print(findingNextPrime(9))
```

```python
# updating region-code values with their next prime value if non_prime

df['region-code'].apply(lambda x:findingNextPrime(x) if not pd.isna(x) else np.nan)
```

In this way through calculating max prime ,,, if no prime exists , then filling by None

```python
# df.apply(lambda x:------)   x is a list of elements of the series elements



def max_prime(li):               # li=list
    mx=float('-inf')
    for x in li:
        if pd.notnull(x):        # only non-null values
            if prime_check(x):
                if x>mx:
                    mx=x
    if mx==float('-inf'):
        return None

    return mx


# finding maximum prime from row elements as LIST

df['NEW_COL2']=df[['country-code','region-code','sub-region-code']].apply(lambda x:max_prime(x),axis=1)


df.head()
```
