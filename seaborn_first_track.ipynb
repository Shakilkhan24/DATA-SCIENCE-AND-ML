```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df.head()
```

Relation_plot


1.   scatterplot [axes]
2.   lineplot [axes]
3.   relplot [figure]



```python
sns.scatterplot(data=df,x='total_bill',y='tip')   # axis level function
```

```python
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter')    # figure level function
                                                              # just putting kind value is enough
                                                              # no need to explicitly calling different plotting functions

                                                              # KIND='your_plot_style'
```

```python
sns.relplot(data=df,x='total_bill',y='tip',kind='line')
```

```python
sns.relplot(data=df,x='total_bill',y='tip',kind='hist')
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip',hue='sex',style='time',size='size')

                                                                      # hue
                                                                      # style
                                                                      # size

                                                                      # KIND OF A MULTI_DIMENSIONAL WAY TO THINK ABOUT
```

```python
sns.relplot(data=df,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')

# complete overview of data multidimensionally
```

```python
sns.lineplot(data=df,x='tip',y='total_bill')   # AXES LEVEL

                                               # hue
                                               # style
                                               # size

# sns.relplot(data=df,x='tip',y='total_bill',kind='lineplot')    # FIGURE LEVEL
```

```python
# facet plot [having col row]
# only for relplot, but not in scatter or lineplot[axes level, no control on figure]

sns.relplot(kind='line',data=df,x='total_bill',y='tip',col='sex')
# col_wrap=4
# one row having only 4 plots for clear view


                                                                 # SPLITTING INTO SUB_PLOTS BASED ON CATEGORIES OF ANY COLUMN

                                                                 # col
                                                                 # row

                                                                 # being alternative to hue,style,size [splitted]

```

CATEGORICAL PLOT....


   SCATTER


1.   __STRIPPLOT
2.   __SWARMPLOT







```python
# strip plot

# x---categorical data
# y--- numerical data
```

```python
sns.stripplot(data=df,x='day',y='total_bill')    # axes leveled
```

```python
sns.catplot(data=df,x='day',y='total_bill',kind='strip',jitter=False)    # figure leveled ,,,cat indicates categorical plot
```

```python
# jitter=True or jitter=True

# check using as an argument

# also can have a hue parameter
sns.catplot(data=df,x='day',y='total_bill',kind='strip',hue='sex')    # x__categorical y__numerical
```

```python
# swarm plot

# kind of the same as strip plot
   # just a little distribution ideas expressing more than stripplot
```

```python
sns.swarmplot(data=df,x='day',y='total_bill')
```
