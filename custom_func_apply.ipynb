````python
import pandas as pd

df=pd.read_csv('https://cdn.wsform.com/wp-content/uploads/2018/09/country_full.csv')
df.head()
````

````python
df.region.value_counts()
````

````python
def func(x,y,z):
  return x=='Americas' and pd.notna(y) and z>=50
mask=df.apply(lambda x:func(x['region'],x['intermediate-region'],x['sub-region-code']),axis=1)   # know the complete setup ...
# axis is mandatory to include____
df[mask].head(2)

# as df.apply() functions' values are a row...so row based indexing is happening here...

# part of optimal filtering ... or a smarter way to manipulte filtering___
````

````python

def contain_words(x,word):
  if pd.isna(x) or not isinstance(x,str):
    return False
  else:
    return x.startswith(word)
mask=df['sub-region'].apply(lambda x:contain_words(x,'Latin'))  # lambda function in apply function is very important...
df[mask].shape[0]
````
